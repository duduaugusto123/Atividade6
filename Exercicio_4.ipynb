```python
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Dados dos pacientes
leucocitos = np.array([2000, 4000, 5000, 6500])
plaquetas = np.array([100000, 20000, 80000, 145000])
linfocitos = np.array([2.3, 4.5, 6.5, 4.4])

# Criar matriz de features
X = np.column_stack((leucocitos, plaquetas, linfocitos))

# Aplicar o algoritmo Isolation Forest
isolation_forest = IsolationForest(contamination=0.25)
labels = isolation_forest.fit_predict(X)

# Identificar anomalias
anomalias_indices = np.where(labels == -1)
anomalias_leucocitos = leucocitos[anomalias_indices]
anomalias_plaquetas = plaquetas[anomalias_indices]
anomalias_linfocitos = linfocitos[anomalias_indices]

# Criar grfico 3D de boxplot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plotar dados normais
ax.scatter(leucocitos, plaquetas, linfocitos, c='b', marker='o', label='Dados Normais')

# Plotar anomalias
ax.scatter(anomalias_leucocitos, anomalias_plaquetas, anomalias_linfocitos, c='r', marker='x', s=100, label='Anomalias')

# Configurar rtulos dos eixos
ax.set_xlabel('Leuccitos')
ax.set_ylabel('Plaquetas')
ax.set_zlabel('Linfcitos')

# Adicionar legenda
ax.legend()

# Exibir o grfico
plt.show()
```
