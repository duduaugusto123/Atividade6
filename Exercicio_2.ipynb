```python
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Criar arrays com os dados
corrente_eletrica = np.array([5, 10, 14, 2, 1.5, 6])
tempo = np.array([1, 2, 4, 6, 7, 10])

# Criar matriz de features
X = np.column_stack((tempo, corrente_eletrica))

# Aplicar o algoritmo Isolation Forest
isolation_forest = IsolationForest(contamination=0.2)
labels = isolation_forest.fit_predict(X)

# Identificar anomalias
anomalias_indices = np.where(labels == -1)
anomalias_corrente = corrente_eletrica[anomalias_indices]
anomalias_tempo = tempo[anomalias_indices]

# Criar gráfico 3D de boxplot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(tempo, corrente_eletrica, c='b', marker='o', label='Dados Normais')
ax.scatter(anomalias_tempo, anomalias_corrente, c='r', marker='x', s=100, label='Anomalias')

ax.set_xlabel('Tempo')
ax.set_ylabel('Corrente Elétrica')
ax.set_zlabel('Valor')
ax.legend()
plt.show()
```
