```python
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Criar arrays com os dados
leucocitos = np.array([2000, 4000, 5000, 6500])
plaquetas = np.array([100000, 20000, 80000, 145000])
linfocitos = np.array([2.3, 4.5, 6.5, 4.4])

# Criar matriz de features
X = np.column_stack((leucocitos, plaquetas, linfocitos))

# Aplicar o algoritmo DBSCAN
dbscan = DBSCAN(eps=2000, min_samples=2)
labels = dbscan.fit_predict(X)

# Criar grfico 3D de boxplot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plotar cada grupo com um boxplot 3D
for label in set(labels):
    indices_grupo = np.where(labels == label)
    ax.scatter(leucocitos[indices_grupo], plaquetas[indices_grupo], linfocitos[indices_grupo], label=f'Grupo {label}')

ax.set_xlabel('Leuccitos')
ax.set_ylabel('Plaquetas')
ax.set_zlabel('Linfcitos')
ax.legend()
plt.show()
```
